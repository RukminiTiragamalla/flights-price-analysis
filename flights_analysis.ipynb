```python
#Showing departure time & Arrival Time for flights with their count
departure_counts = data['departure_time'].value_counts()
arrival_counts=data['arrival_time'].value_counts()
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(departure_counts.index, departure_counts.values, color='blue')
plt.xlabel("Departure Time")
plt.ylabel("Departure frequency")
plt.title("Departure Time Frquency")
plt.subplot(1,2,2)
plt.bar(arrival_counts.index,arrival_counts.values,color='orange')
plt.xticks(rotation=30)
plt.xlabel("Arrival Time")
plt.ylabel("Arrival Frquency")
plt.title("Arrival Time Frquency")
plt.show()
```

```python
#Shwoing the Source city & Destination city of the flights with their Counts
source_city_counts=data['source_city'].value_counts()
destination_city_counts=data['destination_city'].value_counts()
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.barh(source_city_counts.index,source_city_counts.values,color='orange')
plt.xlabel('Frequency Count')
plt.ylabel('Source city')
plt.title('Source City With No.of Flights')
plt.subplot(1,2,2)
plt.barh(destination_city_counts.index,destination_city_counts.values)
plt.xlabel('Frequency Count')
plt.ylabel('Destination City')
plt.title('Destination City With No.of Flights')
plt.show()

```
